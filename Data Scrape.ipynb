```python
# Dependencies
import os
from bs4 import BeautifulSoup as bs
import requests
import pymongo
from splinter import Browser
import pprint
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.catalinajobs_db
collection = db.jobs
```

```python
# URL of page to be scraped
url = 'https://www.indeed.com/jobs?as_and&as_phr&as_any&as_not&as_ttl&as_cmp&jt=all&st&salary&radius=25&l=Catalina%20Island%2C%20CA&fromage=any&limit=10&sort&psf=advsrch&from=advancedsearch&vjk=cb501a684babcc31'
# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'html'
soup = bs(response.content, 'html.parser')
print(soup.prettify()).text
```

```python
results = soup.find(id='mosaic-provider-jobcards')
print(results.prettify())
```

```python
for result in results:
        # Identify and return title of job
        title = result.find('h2', class_='jobTitle')
        
        # Identify and return company name
        company = result.find('pre', class_='companyName')
        
        # Identify and return Job Location
        location = result.find('div', class_='companyLocation')
        
        # Identify and return Job Pay
        pay = result.find('div', class_='attribute_snippet')
        
        # Identify and return Job Type
        jobtype = result.find('div', class_='attribute_snippet')
        
        #Identify and return link to job posting
        link = result.find_all('a',['href'])
        
        print('---------------')
        print(title)
        print(company)
        print(location)
        print(pay)
        print(jobtype)
```
